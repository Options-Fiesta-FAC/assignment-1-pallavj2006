```python
import numpy as np
import matplotlib.pyplot as plt

#function to calculate the option payoffs
def payoffs(S, K, P, o_type, position):
    if o_type == 'call':
        intrinsic = np.maximum(0, S - K)
    elif o_type == 'put':
        intrinsic = np.maximum(0, K - S)

    if position == 'buy':
        payoff = intrinsic - P
    elif position == 'sell':
        payoff = P - intrinsic
    return payoff


```

```python
#input statements and linespace array generstion
K = float(input("Enter strike price (K): "))
P = float(input("Enter premium (P): "))
s_min = K - 50
s_max = K + 50
S = np.linspace(s_min, s_max, 100)
```

```python
#setting up the 4 cases
payoff_buy_call = payoffs(S, K, P, 'call', 'buy')
payoff_sell_call = payoffs(S, K, P, 'call', 'sell')
payoff_buy_put = payoffs(S, K, P, 'put', 'buy')
payoff_sell_put = payoffs(S, K, P, 'put', 'sell')
```

```python
#plotting
plt.figure(figsize=(10, 6))
plt.plot(S, payoff_buy_call, label='Buy call', color='blue')
plt.plot(S, payoff_sell_call, label='Sell call', color='red')
plt.plot(S, payoff_buy_put, label='Buy put', color='green')
plt.plot(S, payoff_sell_put, label='Sell put', color='orange')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(K, color='black', linewidth=0.5, linestyle='-', label='Strike Price (K)')
plt.title('Option payoff diagrams')
plt.xlabel('Spot price at expiry (S)')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(False)
plt.show()
"""
explaining the plot...for buy call we see that till the strike price, our payoffs result in a loss(as much as the premium)....
this is because buy call option contracts are worthless till the spot price doesnt cross the strike price. after that the P/L
slowly picks up as we start eating into the premium and P/L reaches zero when the spot price is strike price+premium.
Exactly opposite to it would be sell call where u make a profit of the premium till the spot price is less than the strike price and 
then slowly u start losing money as price goes above the strike price.

For Buy put the buyer makes profit when the spot price is much lesser than the strike price(spot price< strike price -- premium) and
when spot price reaches the strike price he incurs a loss of the premium paid in its entirety and for further spot prices 
the contract becomes worthless and the buyer takes a hit of the premium.
For sell put, the seller makes a loss when the spot price is much lesser than the strike price (spot price< strike price -- premium) and
when spot prices reaches the strike price he incurs a profit of the premium paid to him which remains constant for further values of
spot prices
"""
```
